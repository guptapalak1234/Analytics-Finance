## Global Minimum Variance Portfolio

#### To find out the weights of global minimum variance portfolio, we applied one trick, where we put the same returns for all assets, and given those same returns, ran the optimiser which will maximise the sharpe ratio for given covariance matrix of the assets. This will give the portfolio with minimum variance to adjust for maximum sharpe ratio.

##### Yellow dot represents equally weighted portfolio for the given assets.
##### Green dot represents maximum sharpe ratio portfolio for the given assets.
##### Blue dot represents Global minimum variance portfolio for the given assets.
