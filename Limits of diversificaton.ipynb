### Limits Of Diversification

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import risk_kit as rk
import pandas as pd
import numpy as np
```

```python
ind_return=rk.get_ind_returns()
ind_nfirms=rk.get_ind_nfirms()
ind_size=rk.get_ind_size()
```

```python
ind_mktcap=ind_nfirms * ind_size
```

```python
ind_mktcap
```

```python
total_mktcap=ind_mktcap.sum(axis="columns")
```

```python
total_mktcap["2007":].plot(figsize=(12,6))
```

```python
total_mktcap
```

```python
ind_capweight=ind_mktcap.divide(total_mktcap, axis="rows")
```

```python
ind_capweight
```

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight["2007":][["Fin","Steel"]].plot()
```

```python
total_market_return=(ind_capweight*ind_return).sum(axis="columns")
total_market_return.plot(figsize=(15,6))
```

```python
total_market_index=rk.drawdown(total_market_return)["Wealth"]
total_market_index.plot(figsize=(15,6),title="Total Market CapWeighted Index 1926-2018")
```

```python
total_market_index["2000":].plot(figsize=(15,6))
```

```python
total_market_index["2000":].plot(figsize=(15,6))
total_market_index["2000":].rolling(window=36).mean().plot(figsize=(15,6))
```

```python
rolling36m_mkt_return=total_market_return.rolling(window=36).aggregate(rk.ann_returns)
rolling36m_mkt_return.plot(figsize=(12,6), label="Trailing 36 month returns", legend="True")
total_market_return.plot(label="Returns", legend="True")
```

### Rolling Correlation 

```python
ts_corr=ind_return.rolling(window=36).corr()
ts_corr.tail(30)
```

```python
ts_corr.index
```

```python
ts_corr.index.names=['date','industry']
ts_corr.tail()
```

```python
ind_tr36corr=ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
ind_tr36corr
```

```python
ind_tr36corr.plot()
```

```python
rolling36m_mkt_return.plot(label="Trailing36 Mo Rets",figsize=(12,6), legend=True)
ind_tr36corr.plot(label="Trailing36 Mo Corr",legend=True, figsize=(12,6), secondary_y=True)
```

```python
rolling36m_mkt_return["2007":].plot(label="Trailing36 Mo Rets",figsize=(12,6), legend=True)
ind_tr36corr["2007":].plot(label="Trailing36 Mo Corr",legend=True, figsize=(12,6), secondary_y=True)
```
