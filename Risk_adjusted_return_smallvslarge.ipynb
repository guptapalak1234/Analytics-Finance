```python
import pandas as pd
import numpy as np
```

```python
returns=pd.read_csv("Portfolios_Formed_on_ME_monthly_EW.csv",header=0, index_col=0,parse_dates=True, na_values=-99.99)
returns
```

```python
returns.head()
```

```python
columns=['Lo 10','Hi 10']
returns=returns[columns]
returns.head()
```

```python
returns=returns/100
returns.head()
```

```python
returns.columns=['Smallcap', 'Largecap']
returns.head()
```

```python
returns.plot()
```

### Standard Deviation of Smallcap and Largecap

```python
Volatility=returns.std()
```

```python
Annualized_vol= Volatility*np.sqrt(12)
Annualized_vol
```

```python
n_months=returns.shape[0]
return_per_month=(returns+1).prod()**(1/n_months)-1
return_per_month
```

### Returns of Smallcap and Largecap

```python
Annualized_return=((return_per_month+1)**12)-1
Annualized_return
```

```python
Annualized_return2=((returns+1).prod())**(12/n_months)-1
Annualized_return2
```

### Sharpe Ratio

```python
risk_free_rate=0.03
excess_return=Annualized_return-risk_free_rate
Sharpe_ratio=excess_return/Annualized_vol
Sharpe_ratio
```
